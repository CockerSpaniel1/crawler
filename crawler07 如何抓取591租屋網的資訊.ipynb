```python
import requests
from bs4 import BeautifulSoup
import json

headers  = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
index = "https://sale.591.com.tw/"

s = requests.Session()
res = s.get(index, headers=headers )

soup = BeautifulSoup(res.text, 'html.parser')
token_item = soup.select_one('meta[name="csrf-token"]')
```

```python
url = "https://sale.591.com.tw/home/search/list-v2?type=2&category=1&shType=list&firstRow=30&totalRows=22620&regionid=17&timestamp=1718728033424&recom_community=1"

headers['X-CSRF-TOKEN'] = token_item['content']
#headers["Cookie"] = "urlJumpIp=8"

res2 = s.get(url, headers=headers )
data = json.loads(res2.text)
print(data)
```
